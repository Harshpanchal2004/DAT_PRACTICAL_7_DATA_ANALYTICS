```python
# Check invalid sales values
invalid_sales = data[data['Sales Volume'] < 0]
print("Invalid Sales Rows:\n", invalid_sales)
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Ensure numeric
data['price'] = pd.to_numeric(data['price'], errors='coerce')
data['Sales Volume'] = pd.to_numeric(data['Sales Volume'], errors='coerce')

# Drop missing
data = data.dropna(subset=['price', 'Sales Volume'])

# Features & target
X = data[['price']]
y = data['Sales Volume']

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction (FIXED)
new_data = pd.DataFrame([[50]], columns=['price'])
prediction = model.predict(new_data)

print("Prediction for price 50:", prediction[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['price'], data['Sales Volume'])
plt.plot(data['price'], model.predict(X))
plt.xlabel("Price")
plt.ylabel("Sales Volume")
plt.title("Price vs Sales Prediction")
plt.show()
```
